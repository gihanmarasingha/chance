# Everybody counts: chance and health

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## The birthday problem
```

```python
def bday(n):
    """The probability that, out of n people, at least two have the same birthday"""
    p = 1
    for i in range(2,n+1):
        p = p * (365 - (i-1))/365
    return 1 - p
```

```python
def bday_plot(n):
    x_values = np.arange(2, n+1)  # Generate integer values from 2 to n
    y_values = [bday(x) for x in x_values]

    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, marker='o', linestyle='-')
    plt.xlabel('Number of people')
    plt.ylabel('Probability')
    plt.title(f'Probability that at least two people in a room of {n} have the same birthday')
    plt.grid(True)
    plt.show()
```

```python
bday_plot(50)
```

## Maximum run length in coin tosses

```python
def max_run_length(arr):
    max_run = 0
    current_run = 1
    for i in range(1, len(arr)):
        if arr[i] == arr[i-1]:
            current_run += 1
        else:
            max_run = max(max_run, current_run)
            current_run = 1
    max_run = max(max_run, current_run)
    return max_run
```

```python
def plot_max_run_histogram(m, n):
    max_run_lengths = []
    for _ in range(m):
        arr = np.random.choice([0, 1], size=n)
        max_run = max_run_length(arr)
        max_run_lengths.append(max_run)

    plt.hist(max_run_lengths, bins=range(min(max_run_lengths), max(max_run_lengths) + 1), align='left')
    plt.xlabel('Maximum Run Length')
    plt.ylabel('Frequency')
    plt.title('Histogram of Maximum Run Lengths')
    plt.show()

    print(f"Mean maximum run length is {np.mean(max_run_lengths)}")
```

```python
plot_max_run_histogram(10, 30)
```
