```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def bday(n):
    """The probability that, out of n people, at least two have the same birthday"""
    p = 1
    for i in range(2,n+1):
        p = p * (365 - (i-1))/365
    return 1 - p
```

```python
# Step 3: Generate Data Points
x_values = np.arange(2, 61)  # Generate integer values from 2 to 60
y_values = [bday(x) for x in x_values]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, marker='o', linestyle='-')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Plot of my_function')
plt.grid(True)
plt.show()
```
